```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# To see the plot inside jupyter notebook
%matplotlib inline
# To see the plot in other code editors and environments use 'plt.show()' after creating the plot
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

### Functional Method

```python
# Creating plots
plt.plot(x, y)
```

```python
# Add labels and title to the plot
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# Multiplots on the same canvas
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

### Object Oriented Method

```python
# Creating plots
# Create a figure object (blank canvas)
fig = plt.figure()
# Add axis to the figure object
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
```

```python
# Add labels and titles to plot
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
# Multiplots on the same canvas
fig = plt.figure()
# Position of the plot [left, bottom, width, height] 
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title('Larger Plot')

axes2.plot(y, x)
axes2.set_title('Smaller Plot')
```

```python
# Multiplots using subplots
fig, axes = plt.subplots(nrows=3, ncols=2)
# Remove overlaps
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
# axes is an array of matplotlib axis objects
for ax in axes:
    ax.plot(x,y)
axes[0].set_title('Plot 1')
axes[1].set_title('Plot 2')
```

### Figure Size and DPI

```python
# Figure size is the aspect ratio of the plot
fig = plt.figure(figsize=(8, 2))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2))
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

### Saving Figures

```python
# Save figures into files using various formats including JPEG, PNG, ...
# Refer to documentation for the full list of available formats
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Title')
ax.plot(x, y)
fig.savefig('../sample-files/fig.png', dpi=200)
```

```python
# Adding Legend to the plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')
ax.legend(loc=(0.1, 0.1))
```

```python
# Another way of adding Legend to the plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')
ax.legend(['X Squared', 'X Cubed'], bbox_to_anchor=(1.0, 1.0), loc='upper left')
```
