# Visualizing Time Series Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as dates
import seaborn as sns

# To make plots interactive 
%matplotlib notebook
sns.set()
```

```python
mcd = pd.read_csv('../sample-files/mcdonalds.csv', index_col='Date', parse_dates=True)
mcd.head()
```

```python
mcd['Adj. Close'].plot()
```

```python
mcd['Adj. Volume'].plot(figsize=(12,4))
```

```python
# Limit x, y
mcd['Adj. Close'].plot(xlim=['2007-01-01', '2009-01-01'], ylim=(20,60), lw=2, ls='--', c='black')
```

```python
idx = mcd.loc['2007-01-01': '2007-05-01'].index
idx
```

```python
stock = mcd.loc['2007-01-01': '2007-05-01']['Adj. Close']
stock
```

```python
# Editing dates
fig, ax = plt.subplots()
ax.plot_date(idx, stock, '-')
ax.yaxis.grid(True)
ax.xaxis.grid(True)
# Locating
ax.xaxis.set_major_locator(dates.MonthLocator())
# Formatting
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n%B-%Y'))
# Locating
ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=0))
# Formatting
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))
fig.autofmt_xdate()
plt.tight_layout()
```
